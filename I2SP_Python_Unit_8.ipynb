# Plotting in Python

This week we will be introducing plotting in Python. We will be using 3 libraries to do this; [Matplotlib](https://matplotlib.org/), [Seaborn](https://seaborn.pydata.org/), and [Pandas](https://pandas.pydata.org/).

# Simple regression

A simple example of a fitting  some different polynomial regression models. We will look next week at how we can do this properly, but for now I am going to do this by hand. Lets start by generating some data generated based on a cubic regression model.

```python
import numpy as np
beta = [1,-1,-0.1,0.1]
x = np.linspace(-5, 5, 200)
y_true = beta[0] * 1 + beta[1] * x + beta[2] * x**2 + beta[3] * x**3
y = y_true + np.random.normal(0,1,200)
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.show()
```

Not quite what we are after, lets make it points

```python
plt.plot(x, y, 'o')
plt.show()
```

Lets add axes and a title, so our plots are clearer

```python
# plot the data
plt.plot(x, y, 'o')
# add axes titles
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Regression')
# show plot
plt.show()
```

Let's make the plot a bit bigger so we can see it better

```python
# Figure size
plt.figure(figsize=[9,6])
# plot the data
plt.plot(x, y, 'o')
# add axes titles
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Regression')
# show plot
plt.show()
```

and lets make the font sixe bigger so we can see it better

```python
# Figure size
plt.figure(figsize=[9,6])
# font size
plt.rcParams.update({'font.size': 20})
# plot the data
plt.plot(x, y, 'o')
# add axes titles
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Regression')
# show plot
plt.show()
```

You can also do this for individual parts of the figure. See the second answer at this [link](https://stackoverflow.com/questions/3899980/how-to-change-the-font-size-on-a-matplotlib-plot)

## Adding lines and legends

Lets add a legend. Note I added both the label and the `plt.legend()` command

```python
# Figure size
plt.figure(figsize=[9,6])
# font size
plt.rcParams.update({'font.size': 20})
# plot the data
plt.plot(x, y, 'o', label='Data Points')
# add axes titles
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Regression')
# add legend
plt.legend()
# show plot
plt.show()
```

Let's add some lines for the different types of regression model. Remember I am just approximating the models here, we should always fit them properly but we will come to that next week.

```python
beta_intercept = [0]
y_intercept_predict = beta_intercept[0] * np.array([1 for i in range(len(x))])
beta_linear = [0, 1]
y_linear_predict = beta_linear[0] * 1 + beta_linear[1] * x
beta_quadratic = [-5, 0, 0.5]
y_quadratic_predict = beta_quadratic[0] * 1 + beta_quadratic[1] * x + beta_quadratic[2] * x**2
beta_cubic = [1,-1,-0.1,0.1]
y_cubic_predict = beta_cubic[0] * 1 + beta_cubic[1] * x + beta_cubic[2] * x**2 + beta_cubic[3] * x**3
```

```python
# Figure size
plt.figure(figsize=[9,6])
# font size
plt.rcParams.update({'font.size': 20})
# plot the data
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x, y_intercept_predict, label='Intercept Model')
plt.plot(x, y_linear_predict, label='Linear Model')
plt.plot(x, y_quadratic_predict, label='Quadratic Model')
plt.plot(x, y_cubic_predict, label='Cubic Model')
# add axes titles
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Regression')
# add legend
plt.legend()
# show plot
plt.show()
```

Lets make the lines a little bigger and lets try to reduce the size of the text in the legend

```python
# Figure size
plt.figure(figsize=[9,6])
# font size
plt.rcParams.update({'font.size': 20})
plt.rc('legend', fontsize=12)
# plot the data
line_width = 4
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x, y_intercept_predict, linewidth=line_width, label='Intercept Model')
plt.plot(x, y_linear_predict, linewidth=line_width, label='Linear Model')
plt.plot(x, y_quadratic_predict, linewidth=line_width, label='Quadratic Model')
plt.plot(x, y_cubic_predict, linewidth=line_width, label='Cubic Model')
# add axes titles
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Regression')
# add legend
plt.legend()
# show plot
plt.show()
```

Lets add a second plot which shows the residuals for the cubic model.

```python
residuals_cubic = y_cubic_predict - y
```

```python
# Figure size
plt.figure(figsize=[18,6])

##### Figure 1 #####

# Select first plot in row of two plots
plt.subplot(1, 2, 1)
# font size
plt.rcParams.update({'font.size': 20})
plt.rc('legend', fontsize=12)
# plot the data
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x, y_intercept_predict, linewidth=4, label='Intercept Model')
plt.plot(x, y_linear_predict, linewidth=4, label='Linear Model')
plt.plot(x, y_quadratic_predict, linewidth=4, label='Quadratic Model')
plt.plot(x, y_cubic_predict, linewidth=4, label='Cubic Model')
# add axes titles
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Regression')
# add legend
plt.legend()

##### Figure 2 #####

# Select second plot in row of two plots
plt.subplot(1, 2, 2)
# plot the residuals
plt.plot(x, residuals_cubic, 'o')
# add axes titles
plt.xlabel('x')
plt.ylabel('Residuals')
plt.title('Residuals for Cubic Regression')
# add a horizontal line
plt.axhline(y=0, color='r', linestyle='-', linewidth=4)

# show plot
plt.show()
```

# Plotting using Seaborn

Seasborn has some nice plotting options and is build on Matplotlib. We can't quite reproduce the plots before using just Seaborn, but we can at least plot the lines which might be useful for other things

```python
import pandas as pd
data =  pd.DataFrame({'x': x,
                      'y_intercept_predict': y_intercept_predict,
                      'y_linear_predict': y_linear_predict,
                      'y_quadratic_predict': y_quadratic_predict,
                      'y_cubic_predict': y_cubic_predict})
```

```python
data.head()
```

```python
data_melted=data.melt(id_vars=["x"], value_vars=["y_intercept_predict","y_linear_predict","y_quadratic_predict","y_cubic_predict"],
var_name="Model", value_name="y")
data_melted.head()
```

```python
import seaborn as sns
sns.relplot("x", "y", hue="Model", data=data_melted, kind="line")
plt.show()
```

# Plotting using Pandas

Pandas has a nice plotting interface which can be helpful for quickly visualising some of your data. Lets load some data from a previous week and try some of the options out

```python
url = 'https://raw.githubusercontent.com/UofGAnalyticsData/DPIP/main/colab_datasets/premierleague.csv'
football = pd.read_csv(url)
football = football.set_index('Team')
football.head()
```

You can easily do things like boxplots, barplots and histograms. This can be really useful when you first get a new dataset and want to do some data exploration

```python
football.boxplot(column = ['Pts','GD'])
plt.show()
```

```python
football.hist(column=['GD'])
plt.show()
```

```python
football.plot.bar(y='Pts')
plt.show()
```
