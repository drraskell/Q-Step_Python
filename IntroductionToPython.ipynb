## Python Functions

Code in python functions use indents instead of braces {} (or other structures). This makes the code look cleaner and is usually easier to use but it is sometimes causes some issues.

One old discussion in python programming is the mixture between "tabs" for indent and "spaces" mostly editors just deal with this, but mixing them can cause all sorts of issues!


```python
def fib(num):
  if num==0:
    return 0
  elif num==1:
    return 1
  elif num==2:
    return 1
  else:
    return fib(num-1)+fib(num-2)
```

```python
print(fib(10))
```

### Accessible machine learning libraries
Python also has some very nice machine learning libraries which we will learn about in the final weeks of the course.

Below is an example of training a random forest using Sklearn.

Credit: https://www.blopig.com/blog/2017/07/using-random-forests-in-python-with-scikit-learn/

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
iris
```

Lets first create the model object:

```python
rf = RandomForestClassifier()
```

Then assemble the data into X and Y:

```python
X = iris[[x for x in iris if x!='Species']]
Y = iris['Species']
```

Then fitting the model is a single command:

```python
rf.fit(X,Y)
```

Predicting is equally straightforward

```python
rf.predict(X)
```

Lets look at the insample confusion matrix:

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(rf.predict(X),Y),annot=True)
```

If you would like something more interpretable you could also use a decision tree:

```python
from sklearn import tree
treeC = tree.DecisionTreeClassifier()
treeC = treeC.fit(X,Y)
```

Note the syntax for fitting this model is almost identical. Again lets look at the (insample) predictions:

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(treeC.predict(X),Y),annot=True)
```

```python
from matplotlib import pylab
pylab.figure(figsize=(15,15))
res = tree.plot_tree(treeC,fontsize=12)
```
